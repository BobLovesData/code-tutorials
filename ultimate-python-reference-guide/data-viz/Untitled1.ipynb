```python
import pandas as pd
import calendar as c
c.setfirstweekday(calendar.SUNDAY)
```

```python
start='2022-03-01'
end='2022-04-30'
```

```python
df = pd.DataFrame({"Date": pd.date_range(start, end)})
df['DateCK'] = df.Date.dt.year.astype(str) + df.Date.dt.month.astype(str).str.zfill(2) + df.Date.dt.day.astype(str).str.zfill(2)
df['FormattedDate'] = df.Date.dt.strftime('%m/%d/%Y')
df['DayOfMonth'] = df.Date.dt.day
df['DaySuffix'] = df.Date.dt.day.astype(str) + df.Date.dt.day.map(lambda x: 'th' if 11<=x<=13 else {1:'st',2:'nd',3:'rd'}.get(x%10, 'th'))
df["DayName"] = df.Date.dt.day_name()
df["DayOfWeek"] = df.Date.dt.weekday.map(pd.Series('2,3,4,5,6,7,1'.split(',')))
#df["DayOfWeekInMonth"] = df.Date.map(lambda x: idx[0] for if y == x.day idx, y in np.ndenumerate(calendar.monthcalendar(x.year,x.month)))
#df["DayOfWeekInMonth"] = 
#df["Week"] = df.Date.dt.weekofyear
#df['MonthName'] = df.Date.dt.month.map(lambda x: c.month_name(x))
df['MonthName'] = df.Date.dt.month.map(pd.Series('NONE,January,February,March,April,May,June,July,August,September,October,November,December'.split(',')))
#df['MonthName'] = df.Date.dt.month
#df["Quarter"] = df.Date.dt.quarter
df["Year"] = df.Date.dt.year

df.head(60)
```
